```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Diagramas de Dispersion (Scatterplot)

from turtle import color


x = np.arange(2, 20)
y = x**2
z = x**3

plt.scatter(x, y, color='black', marker='+', label='x^2')
plt.scatter(x, z, color='green', marker='s', label='x^3') 
plt.plot(x, z, color='pink', label='line')

plt.title('Funciones polinómicas', fontsize=20, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()

plt.text(5, 4000, 'texto', color='blue')
```
